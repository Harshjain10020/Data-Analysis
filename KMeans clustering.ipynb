```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd


scaler = MinMaxScaler()


model = scaler.fit(df.drop(['StudentID', 'Gender', 'Ethnicity', 'ParentalEducation', 'Tutoring', 'ParentalSupport', 'Extracurricular', 'Sports', 'Music', 'Volunteering', 'GradeClass'], axis=1))


df[['Age', 'StudyTimeWeekly', 'Absences', 'GPA']] = model.transform(df.drop(['StudentID', 'Gender', 'Ethnicity', 'ParentalEducation', 'Tutoring', 'ParentalSupport', 'Extracurricular', 'Sports', 'Music', 'Volunteering', 'GradeClass'], axis=1))


print(df.head())
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


wcss = []


numeric_df = df.drop(['StudentID', 'Gender', 'Ethnicity', 'ParentalEducation', 'Tutoring', 'ParentalSupport', 'Extracurricular', 'Sports', 'Music', 'Volunteering', 'GradeClass'], axis=1)


for i in range(1, 10):
    kmeans = KMeans(n_clusters=i, random_state=3)
    kmeans.fit(numeric_df)  # Fit the model on the numeric data
    wcss.append(kmeans.inertia_)  # Append the within-cluster sum of squares (WCSS) for each cluster count


plt.plot(range(1, 10), wcss, '-o')
plt.xlabel('No. of clusters')
plt.ylabel('WCSS')
plt.show()

print('wcss: ' + str(wcss))
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score


l1, l2, l3, l4 = [], [], [], []


numeric_df = df.drop(['StudentID', 'Gender', 'Ethnicity', 'ParentalEducation', 'Tutoring', 'ParentalSupport', 'Extracurricular', 'Sports', 'Music', 'Volunteering', 'GradeClass'], axis=1)


for i in (2, 3, 4, 5):
    kmean = KMeans(n_clusters=i, random_state=42)
    mod_k = kmean.fit(numeric_df)  # Fit the model on numeric data
    pred_k1 = mod_k.predict(numeric_df)  # Predict cluster labels

   
    kmean_sil = silhouette_score(numeric_df, pred_k1, metric='euclidean')
    kmean_cal = calinski_harabasz_score(numeric_df, pred_k1)
    kmean_dav = davies_bouldin_score(numeric_df, pred_k1)

    
    print(i, '\t', 'Silhouette Score:', kmean_sil, ', Calinski Harbasz Score:', kmean_cal, ', Davies Bouldin Score:', kmean_dav)
    
    
    l1.append(kmean_sil)
    l2.append(kmean_cal)
    l3.append(kmean_dav)
    l4.append(i)
```

```python
import pandas as pd


results_df = pd.DataFrame({
    'Cluster': l4,
    'Silhouette Score': l1,
    'Calinski Harabasz Index': l2,
    'Davies Bouldin Score': l3
})


print(results_df)
```
