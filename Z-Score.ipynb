```python
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler
data = np.array([10, 12, 14, 15, 18, 20, 21, 22, 23, 24, 25, 26, 28, 30, 32]).reshape(-1, 1)
scaler_z = StandardScaler()
z_score_scaled = scaler_z.fit_transform(data)
scaler_minmax = MinMaxScaler()
minmax_scaled = scaler_minmax.fit_transform(data)
print("Original Data:\n", data.flatten())
print("Z-Score Scaled Data:\n", z_score_scaled.flatten())
print("Min-Max Scaled Data:\n", minmax_scaled.flatten())
```
